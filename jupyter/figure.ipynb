```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
x = np.random.normal(math.pi * 17.5 * 13.5,(math.pi * 20 * 15 - math.pi * 15 * 12)/2, 250)
```

```python
print(x)
```

```python
plt.figure(dpi=1200)

plt.hist(x, 50)

yint = []
locs, labels = plt.yticks()
for each in locs:
    yint.append(int(each))
plt.yticks(yint)

# only one line may be specified; full height
plt.axvline(x = 500, color = 'b', label = 'Ngưỡng dưới - 500 cm2')

plt.axvline(x = 1000, color = 'r', label = 'Ngưỡng trên - 1000 cm2')
 
# place legend outside
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'best')

plt.title('Phân bố kích thước thanh long')
plt.xlabel('Kích thước (cm2)')
plt.ylabel('Số lượng quả')

plt.savefig('size.png', bbox_inches="tight")

plt.show() 
```
